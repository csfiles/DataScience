```python
NA = (spotify.isna().sum()/len(spotify))*100
NA = pd.DataFrame(NA, columns = ['Percentage of Missing Values'])
NA.index.name = 'Name'
plt.figure(figsize=(16,6))
sns.barplot(x='Name',y='Percentage of Missing Values',data = NA)
plt.xticks(rotation=45)
```

```python
NA = (spotify.isna().sum()/len(spotify))*100
NA = pd.DataFrame(NA, columns = ['Percentage of Missing Values'])
NA.index.name = 'Name'
plt.figure(figsize=(16,6))
sns.barplot(x='Name',y='Percentage of Missing Values',data = NA, palette = "deep")
plt.xticks(rotation=45)
```

```python
NA = (spotify.isna().sum()/len(spotify))*100
NA = pd.DataFrame(NA, columns = ['Percentage of Missing Values'])
NA.index.name = 'Name'
plt.figure(figsize=(12,6))
colors = ['red', 'blue', 'green', 'purple']
sns.barplot(x='Name',y='Percentage of Missing Values',data = NA, palette = colors)
plt.xticks(rotation=45)
```

```python
NA = (spotify.isna().sum()/len(spotify))*100
NA = pd.DataFrame(NA, columns = ['Percentage of Missing Values'])
NA.index.name = 'Name'
plt.figure(figsize=(12,6))
colors = ['red', 'blue', 'green', 'purple']
barplt = sns.barplot(x='Name',y='Percentage of Missing Values',data = NA, palette = colors)
for bar in barplt.patches:
  barplt.annotate(format(bar.get_height(), '.2f'),
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), ha='center',va = 'center',size = 10,xytext = (0,5),textcoords = 'offset points')
plt.xticks(rotation=45)
```
